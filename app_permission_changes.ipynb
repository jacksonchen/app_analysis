```python
import pandas as pd
import subprocess
import os
import re
import datetime
import time
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import HTML
```

```python
app_dump = pd.read_csv('/Users/Jackson/dev/science/health-apps/10-health-apps-additional_info.csv')
app_dump.sort(columns=['package', 'version_code'], inplace=True)
app_dump.set_index(['package','version_code'], inplace=True)
app_dump
```

```python
permissions_list = []
df_result = pd.DataFrame(columns=['package', 'verc'])
counter = 2
for gp in app_dump.index.get_level_values(app_dump.index.names[0]).unique():
    for perms in app_dump.loc[gp, "permissions"]:
        permissions_sublist = perms.split(',')
        for permission in permissions_sublist:
            if (permission not in permissions_list) and (re.match(r"android.permission.+" ,permission)):
                permissions_list.append(permission)
                df_result.insert(counter, permission, int)
                counter += 1

df_result.set_index(['package', 'verc'], inplace=True)
counter2 = 0
for gp in app_dump.index.get_level_values(app_dump.index.names[0]).unique():
    for vc in app_dump.iloc[app_dump.index.get_level_values('package') == gp].index.get_level_values(app_dump.index.names[1]).unique():
        #df_result.ix[(gp, vc),:]= "test"
        permscheck_sublist = app_dump.loc[(gp, vc), "permissions"].split(',')
        #stuff = df_result.loc[(gp, vc),:] = "test"
        overall_perms = df_result.columns.values
        for perm in overall_perms:
            if perm in permscheck_sublist:
                df_result.loc[(gp, vc),perm] = 1
            else:
                df_result.loc[(gp, vc),perm] = 0
df_result
# f = open('temp.html', 'wb')
# f.write(df_result.to_html())
# f.close()
```

```python
#This cell is used to calculate the overall statistics of the apps

#Most, least, average permissions calculation
permissions_count = pd.DataFrame(columns = ["package", "versions", "permissions_count"])
permissions_count.set_index(['package', 'versions'], inplace=True)
for gp in df_result.index.get_level_values(df_result.index.names[0]).unique():
    for vc in df_result.iloc[df_result.index.get_level_values('package') == gp].index.get_level_values(df_result.index.names[1]).unique():
        perm_sum = 0
        for perm in df_result.loc[gp, vc]:
            if perm == 1:
                perm_sum += 1
        permissions_count.loc[(gp, vc), "permissions_count"] = perm_sum
# print "max", permissions_count.max()
# print "min", permissions_count.min()
# print "average", permissions_count.mean()

#Which permissions are used the most/least?
permissions_occurence = pd.DataFrame(index = ["permission"], columns = ["occurences"])
for perm in df_result.columns:
    total_app_perms = 0
    for gp in df_result.index.get_level_values(df_result.index.names[0]).unique():
        temp_perm_counter = []
        counter = 0
        for vc in df_result.iloc[df_result.index.get_level_values('package') == gp].index.get_level_values(df_result.index.names[1]).unique():
            temp_perm_counter.append(df_result.ix[[gp, vc], perm][counter])
            counter += 1
        total_app_perms += max(temp_perm_counter)
    permissions_occurence.loc[perm] = total_app_perms
#print "max", permissions_occurence.max()

#How does the number of permissions correlate with app popularity?
permissions_download = pd.DataFrame(columns = ["permissions", "downloads"])
permissions_rating = pd.DataFrame(columns = ["permissions", "rating"])
counter3 = 0
for gp in df_result.index.get_level_values(df_result.index.names[0]).unique():
    for vc in df_result.iloc[df_result.index.get_level_values('package') == gp].index.get_level_values(df_result.index.names[1]).unique():
        #permissions_download.loc[permissions_count.loc[(gp, vc), "permissions_count"]] = app_dump.loc[(gp, vc), "download"]
        permissions_download.loc[counter3] = [permissions_count.loc[(gp, vc), "permissions_count"], app_dump.loc[(gp, vc), "download"]]
        permissions_rating.loc[counter3] = [permissions_count.loc[(gp, vc), "permissions_count"], app_dump.loc[(gp, vc), "rating"]]
        counter3 += 1
permissions_download.sort(columns=["permissions"], inplace=True)
permissions_rating.sort(columns=["permissions"], inplace=True)

#permissions_download.plot(x="permissions", y="downloads", kind="scatter")
permissions_rating.plot(x="permissions", y="rating", kind="scatter")

#permissions_rating.set_index('permissions', inplace=True)
#permissions_download.set_index('permissions', inplace=True)


#f = open('perms_stats.html', 'wb')
# f.write(permissions_count.to_html())
# f.write(permissions_occurence.to_html())
# f.write("How does this relate to app popularity?")
# f.write(permissions_download.to_html())
#f.write(permissions_rating.to_html())
#f.close()
```
