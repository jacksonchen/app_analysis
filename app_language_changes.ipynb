```python
import pandas as pd
import matplotlib
import re
from IPython.display import HTML
```

```python
app_language_dump = pd.read_csv('/Users/Jackson/dev/science/health-apps/locales.csv')
app_language_dump.columns = ['package', 'verc', 'num_locales', 'locales_list']
app_language_dump.sort(columns=['package', 'verc'], inplace=True)
app_language_dump.set_index(['package','verc'], inplace=True)
app_language_dump
```

```python
#Max, min, and average number of supported languages

f = open('locale_stats.html', 'wb')
app_locale_occurence = app_language_dump.sort(columns=['num_locales']).copy(deep=True)
app_locale_occurence.drop('locales_list', axis=1, inplace=True)
temp = app_locale_occurence.copy(deep=True)
f.write(app_locale_occurence.to_html())

print "max", app_locale_occurence.max()
print "min", app_locale_occurence.min()
print "avg", app_locale_occurence.mean()


#Determines all the indices with max value
for i in temp.index:
    if temp.max()[0] == app_locale_occurence.max()[0]:
        print "maximum id", temp.idxmax()
        temp.drop(temp.idxmax(), axis=0, inplace=True)
    else:
        break

# print "================"

more_20_locale = app_locale_occurence.copy(deep=True)
more_20_locale = more_20_locale[more_20_locale.num_locales > 20]

more_50_locale = more_20_locale.copy(deep=True)
more_50_locale = more_50_locale[more_50_locale.num_locales > 50]

more_80_locale = more_50_locale.copy(deep=True)
more_80_locale = more_80_locale[more_80_locale.num_locales > 80]

#more_20_locale
#more_50_locale.shape
#more_80_locale
f.write("App versions with greater than 20 locales: " + str(more_20_locale.shape[0]))
f.write(more_20_locale.to_html())
f.write("App versions with greater than 50 locales: " + str(more_50_locale.shape[0]))
f.write(more_50_locale.to_html())
f.write("App versions with greater than 80 locales: " + str(more_80_locale.shape[0]))
f.write(more_80_locale.to_html())
f.close()
```

```python
#Most commonly supported language

locale_occurence = pd.DataFrame(columns = ['occurence'])
for gp in app_language_dump.index.get_level_values(app_language_dump.index.names[0]).unique():
    for vc in app_language_dump.iloc[app_language_dump.index.get_level_values(app_language_dump.index.names[0]) == gp].index.get_level_values(app_language_dump.index.names[1]).unique():
        m = re.search('--_--(.+)', app_language_dump.loc[(gp, vc), 'locales_list'])
        for locale in m.groups(0)[0].split('|'):
            if locale != "":
                if locale in locale_occurence.index:
                    locale_occurence.loc[locale] += 1
                else:
                    locale_occurence.loc[locale] = 1
locale_occurence.sort(columns=['occurence'], ascending=False, inplace=True)
locale_occurence.index.name = "locale"
f = open('locale_stats.html', 'wb')
f.write(locale_occurence.to_html())
f.close()
```

```python
#Compiles dataframe containing 0s and 1s for if each app version contains each language code

locale_version = pd.DataFrame(columns=['package', 'verc'])
locale_version.set_index(['package', 'verc'], inplace=True)
#Setting the data in the locale_version dataframe
for gp in app_language_dump.index.get_level_values(app_language_dump.index.names[0]).unique():
    for vc in app_language_dump.iloc[app_language_dump.index.get_level_values(app_language_dump.index.names[0]) == gp].index.get_level_values(app_language_dump.index.names[1]).unique():
        m = re.search('--_--(.+)', app_language_dump.loc[(gp, vc), 'locales_list'])
        for locale in m.groups(0)[0].split('|'):
            if locale != "":
                locale_version.loc[(gp, vc), locale] = 1

#Sets all NaNs to 0
for gp in locale_version.index.get_level_values(locale_version.index.names[0]).unique():
    for vc in locale_version.iloc[locale_version.index.get_level_values(locale_version.index.names[0]) == gp].index.get_level_values(locale_version.index.names[1]).unique():
        for lc in locale_version.columns:
            if locale_version.loc[(gp, vc), lc] != 1:
                locale_version.loc[(gp, vc), lc] = 0

locale_version
```

```python
#Finds changes in language support between versions
lc_added = pd.DataFrame(columns=['additions'])
lc_added.index.name = "language"
app_added = pd.DataFrame(columns=['additions'])
app_added.index.name = "app"

lc_removed = pd.DataFrame(columns=['removals'])
lc_removed.index.name = "language"
app_removed = pd.DataFrame(columns=['removals'])
app_removed.index.name = "app"

for gp in locale_version.index.get_level_values(locale_version.index.names[0]).unique():            
    for lc in locale_version.columns:
        before = -1
        for vc in locale_version.iloc[locale_version.index.get_level_values(locale_version.index.names[0]) == gp].index.get_level_values(locale_version.index.names[1]).unique():
            if before == -1:
                before = locale_version.loc[(gp, vc), lc]
            else:
                if locale_version.loc[(gp, vc), lc] - before == -1:
                    if lc in lc_removed.index:
                        lc_removed.loc[lc] += 1
                    else:
                        lc_removed.loc[lc] = 1
                    if gp in app_removed.index:
                        app_removed.loc[gp] += 1
                    else:
                        app_removed.loc[gp] = 1
                elif locale_version.loc[(gp, vc), lc] - before == 1:
                    if lc in lc_added.index:
                        lc_added.loc[lc] += 1
                    else:
                        lc_added.loc[lc] = 1
                    if gp in app_added.index:
                        app_added.loc[gp] += 1
                    else:
                        app_added.loc[gp] = 1

f = open('locale_versions.html', 'wb')
f.write(lc_added.to_html())
f.write(app_added.to_html())
f.write(lc_removed.to_html())
f.write(app_removed.to_html())
f.close()
```
