```python
# Plots the number of supported screen sizes with the download count

sizes_downloads = pd.DataFrame(columns = ['package', 'verc', 'sizes', 'downloads'])
sizes_downloads.set_index(['package', 'verc'], inplace=True)
for gp in app_screen_size_dump.index.get_level_values(app_screen_size_dump.index.names[0]).unique():
    for vc in app_screen_size_dump.iloc[app_screen_size_dump.index.get_level_values(app_screen_size_dump.index.names[0]) == gp].index.get_level_values(app_screen_size_dump.index.names[1]).unique():
        sizes_downloads.loc[(gp, vc), 'sizes'] = app_screen_size_dump.loc[(gp, vc), 'num_sizes']
        sizes_downloads.loc[(gp, vc), 'downloads'] = app_dump.loc[(gp, vc), 'download']

sizes = sizes_downloads['sizes'].tolist()
downloads = sizes_downloads['downloads'].tolist()

fit = polyfit(sizes, downloads, 1)
fit_fn = poly1d(fit)
plot(sizes, downloads, "bo", sizes, fit_fn(sizes), "--k")
ylim(4000000, 11000000)
xlim(2.8, 4.2)
xlabel("sizes")
ylabel("downloads (10 millions)")

res = sm.OLS(downloads, sizes)
print res.fit().summary()
```
