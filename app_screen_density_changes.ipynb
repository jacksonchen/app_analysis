```python
import pandas as pd
import matplotlib
import re
from IPython.display import HTML
```

```python
app_screen_density_dump = pd.read_csv('/Users/Jackson/dev/science/health-apps/screen_densities.csv')
app_screen_density_dump.columns = ['package', 'verc', 'num_densities', 'densities_list']
app_screen_density_dump.sort(columns=['package', 'verc'], inplace=True)
app_screen_density_dump.set_index(['package','verc'], inplace=True)
app_screen_density_dump
```

```python
#Finds the max, min, and avg number of screen densities supported

app_density_occurence = app_screen_density_dump.sort(columns=['num_densities']).copy(deep=True)
app_density_occurence.drop('densities_list', axis=1, inplace=True)
#print app_density_occurence

print "max", app_density_occurence.max()
print "min", app_density_occurence.min()
print "avg", app_density_occurence.mean()

f = open('densities_stats.html', 'wb')              
f.write(app_density_occurence.to_html())
f.close()
```

```python
# Which screen density was most commonly supported?

screen_densities = pd.DataFrame(columns=['occurences'])
screen_densities.index.name = "screen_densities"
for gp in app_screen_density_dump.index.get_level_values(app_screen_density_dump.index.names[0]).unique():
    for vc in app_screen_density_dump.iloc[app_screen_density_dump.index.get_level_values(app_screen_density_dump.index.names[0]) == gp].index.get_level_values(app_screen_density_dump.index.names[1]).unique():
        for density in app_screen_density_dump.loc[(gp, vc), 'densities_list'].split():
            if density not in screen_densities.index:
                screen_densities.loc[density] = 1
            else:
                screen_densities.loc[density] += 1

# f = open('densities_stats.html', 'wb')              
# f.write(screen_densities.to_html())
# f.close()
screen_densities.max()
```

```python
density_version = pd.DataFrame(columns=['package', 'verc'])

# Inputs the data (1s) into the density_version dataframe 
density_version.set_index(['package', 'verc'], inplace=True)
for gp in app_screen_density_dump.index.get_level_values(app_screen_density_dump.index.names[0]).unique():
    for vc in app_screen_density_dump.iloc[app_screen_density_dump.index.get_level_values(app_screen_density_dump.index.names[0]) == gp].index.get_level_values(app_screen_density_dump.index.names[1]).unique():
        for density in app_screen_density_dump.loc[(gp, vc), 'densities_list'].split():
            density_version.loc[(gp, vc), density] = 1
            
#Changes all NANs to 0's in the density_version dataframe
for gp in density_version.index.get_level_values(density_version.index.names[0]).unique():
    for vc in density_version.iloc[density_version.index.get_level_values(density_version.index.names[0]) == gp].index.get_level_values(density_version.index.names[1]).unique():
        for density in density_version.columns:
            if density_version.loc[(gp, vc), density] != 1:
                density_version.loc[(gp, vc), density] = 0
density_version
```

```python
# Which screen densities are most commonly added or removed between app versions?
# Which apps removed/added the most amount of screen densities between versions?

den_added = pd.DataFrame(columns=['additions'])
den_added.index.name = "screen_density"
app_added = pd.DataFrame(columns=['additions'])
app_added.index.name = "app"

den_removed = pd.DataFrame(columns=['removals'])
den_removed.index.name = "screen_density"
app_removed = pd.DataFrame(columns=['removals'])
app_removed.index.name = "app"


for gp in density_version.index.get_level_values(density_version.index.names[0]).unique():
    for density in density_version.columns:
        before = -1
        for vc in density_version.iloc[density_version.index.get_level_values(density_version.index.names[0]) == gp].index.get_level_values(density_version.index.names[1]).unique():
            if before == -1:
                before = density_version.loc[(gp, vc), density]
            else:
                if density_version.loc[(gp, vc), density] - before == -1:
                    if density in den_removed.index:
                        den_removed.loc[density] += 1
                    else:
                        den_removed.loc[density] = 1
                    if gp in app_removed.index:
                        app_removed.loc[gp] += 1
                    else:
                        app_removed.loc[gp] = 1
                elif density_version.loc[(gp, vc), density] - before == 1:
                    if density in den_added.index:
                        den_added.loc[density] += 1
                    else:
                        den_added.loc[density] = 1
                    if gp in app_added.index:
                        app_added.loc[gp] += 1
                    else:
                        app_added.loc[gp] = 1
f = open('densities_versions.html', 'wb')              
f.write(den_added.to_html())
f.write(app_added.to_html())
f.write(den_removed.to_html())
f.write(app_removed.to_html())
f.close()
```
