```python
def get_time_diff(dataframe, old_column, date_column_name):
    # Get the first index name (package name)
    first_index = dataframe.index.names[0]
    df_result = pd.DataFrame(columns=[first_index])
    df_result.set_index(first_index, inplace =True)
    for gp in dataframe.index.get_level_values(first_index).unique():
        tmp_pmdf = dataframe.loc[gp, old_column]
        dates = []
        for r in tmp_pmdf:
            dates.append(datetime.datetime.strptime(r, "%Y-%m-%d"))
        #print (dates[1]-dates[0]).days
        #break
        df_result.ix[gp, date_column_name] = (dates[1]-dates[0]).days
    return df_result
```

```python
df_time_changes = get_time_diff(df_time_analysis, "date_published", "date_diff")
f = open('date_change.html', 'wb')
f.write(df_time_changes.to_html())
f.close()
```

```python
df_permissions = pd.DataFrame(data=apps, columns=["package", "verc", "version_name", "total_permissions", "permissions"])
df_permissions.sort(columns=['package', 'verc'], inplace=True)
df_permissions.set_index(['package','verc'], inplace=True)
df_permissions
```

```python
def get_diff(dataframe, permission_column_name, permission_result_column_name):
    # Get the first index name (package name)
    first_index = dataframe.index.names[0]
    df_result = pd.DataFrame(columns=[first_index, 'number_of_versions', permission_result_column_name])
    df_result.set_index(first_index, inplace =True)
    for gp in dataframe.index.get_level_values(first_index).unique():
        tmp_pmdf = dataframe.loc[gp, permission_column_name]
        perm = 0
        if len(tmp_pmdf) == 1:
            perm = 0
        else:
            for r in tmp_pmdf:
                perm = r - perm
        df_result.ix[gp, permission_result_column_name] = perm
        df_result.ix[gp, 'number_of_versions'] = len(tmp_pmdf)
    return df_result
```

```python
df_permissions_changes = get_diff(df_permissions, 'total_permissions', 'permission_changes')
df_permissions_changes.sort(columns=['permission_changes'], inplace=True)
df_permissions_changes
```
